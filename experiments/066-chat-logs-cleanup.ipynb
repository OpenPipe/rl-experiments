```python
import os
import glob
from pathlib import Path

# Loop through iterations 0001 through 0117
for iteration in range(1, 156):
    iteration_str = f"{iteration:04d}"
    log_dir = f"./models/060/{iteration_str}/chat-completion-logs"
    
    # Check if directory exists
    if not os.path.exists(log_dir):
        print(f"Directory {log_dir} does not exist, skipping")
        continue
        
    # Get all files in the specified directory
    log_files = glob.glob(os.path.join(log_dir, "*"))
    
    # Sort the files
    sorted_log_files = sorted(log_files)
    
    # Keep only the last 8 files, delete the rest
    files_to_keep = sorted_log_files[-8:] if len(sorted_log_files) > 8 else sorted_log_files
    files_to_delete = [f for f in sorted_log_files if f not in files_to_keep]
    
    # Delete the files
    if files_to_delete:
        print(f"Deleting {len(files_to_delete)} log files in {log_dir}")
        for file in files_to_delete:
            os.remove(file)
            print(f"  Deleted {Path(file).name}")
    else:
        print(f"No files to delete in {log_dir} (less than 8 files found)")
    
    # Display the remaining files
    print(f"\nRemaining files in {log_dir}:")
    for file in files_to_keep:
        print(f"  {Path(file).name}")
    
    print("-" * 50)
```
